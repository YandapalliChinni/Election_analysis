```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the Election Commission of India's main results page
url="https://results.eci.gov.in/PcResultGenJune2024/index.htm"

# Send a GET request to the main results page
res=requests.get(url)

# Check if the request was successful (status code 200)
if res.status_code==200:
    # Parse the HTML content using BeautifulSoup
    x=BeautifulSoup(res.content, "html.parser")

    #Find specific elements containing election results data
    y=x.find("div", class_="table-responsive")

    if y:
        # Extracting data from the result div
        tables = y.find_all("table")

        # Prepare a list to hold the extracted data
        all_data = []

        for i in tables:
            # Process each table to extract relevant data
            rows = i.find_all("tr")
            for j in rows:
                # Extract data from each row (adjust based on actual structure)
                cells = j.find_all("td")
                if cells:
                    # Extract data from cells and add to all_data list
                    row_data = [k.get_text(strip=True) for k in cells]
                    all_data.append(row_data)

        # Define the CSV file name
        csv_file = 'election_results.csv'

        # Define the headers based on your data structure
        headers = ["Party", "Won", "Leading", "Total"]  # Adjust headers based on your data

        # Write data to CSV
        with open(csv_file, 'w', newline='') as file:
            writer = csv.writer(file)
            # Write the header row
            writer.writerow(headers)
            # Write the data rows
            writer.writerows(all_data)

        print(f"Data successfully saved to {csv_file}")
    else:
        print("No election results data found on the main page.")

else:
    print("Failed to retrieve the webpage. Status code:", res.status_code)
```
